```python
df = gpd.read_file(libpysal.examples.get_path('south.shp'))
```

```python
hr60_q10 = mc.Quantiles(df['HR60'], k=10)
hr60_q10
```

```python
hr60_std=mc.Std_Mean(df['HR60'], multiples = [-3, -1.5, 1.5, 3])
hr60_std
```

```python
df=df.assign(color_level=hr60_std.yb)
```

```python
df.plot(column='color_level', categorical=True, k=10, cmap='OrRd', linewidth=0.1, edgecolor='white', legend=True)
```
